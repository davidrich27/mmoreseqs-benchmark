```python
for data in datas:
    data["roc_y"] = data["tot_T"] / total_TP
    data["roc_x"] = data["tot_F"] / total_query
```

```python
# find first false positive
auc_ffps = []

for data in datas:
    ffp = data[ data["tot_F"] > 0 ].iloc[0]["roc_y"]
    auc_ffps.append(ffp)
    
print(auc_ffps)
```

```python
# 1% false recovery
auc_fdrs = []

for data in datas:
    fdr = data[ (data["tot_F"]/data["tot_T"]) > 0.001 ].iloc[0]["roc_y"]
    auc_fdrs.append(fdr)
    
print(auc_fdrs)
```
