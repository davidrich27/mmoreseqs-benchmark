#### MMSEQS+ post-fb-pruner data

```python
# set header names and datatype
mmseqs_plus_dformat = {
    "res_id": "int64", 
    "target_id": "int64",
    "target_len": "int64",
    "query_id": "int64",
    "query_len": "int64",
    "total_cells": "int64",
    "calc_cells": "int64",
    "viterbi_sc": "float64",
    "fb_prune_sc": "float64",
    "forward_sc": "float64",
    "tf": "str"
}
mmseqs_plus_header = mmseqs_plus_dformat.keys()
```

```python
# load data
sep = " "
usecols = range(len(mmseqs_plus_header))
skiprows = []
names = mmseqs_plus_header
mmseqs_plus_data = pd.read_csv( mmseqs_plus_fname, sep=" ", header=None, skiprows=skiprows, usecols = usecols, names=names, index_col=False, verbose=False )
data = mmseqs_plus_data
mmseqs_plus_data
```

```python
print( len(data) )
```

```python
# cast data to proper datatypes
cols = list(mmseqs_plus_data.columns.values)
for i in range(len(cols)):
    col = cols[i]
    dtype = mmseqs_plus_dformat[col]
    mmseqs_plus_data.iloc[:,i] = mmseqs_plus_data.iloc[:,i].astype(dtype)
mmseqs_plus_data.dtypes
```

```python
# add difference
data["perc_cells"] = data.calc_cells / data.total_cells
data["prune_below_fwd"] = data.forward_sc - data.fb_prune_sc
data["vit_below_fwd"] = data.forward_sc - data.viterbi_sc
data
```

```python
trues = data[data.tf == "T"]
trues
falses = data[data.tf == "F"]
falses
```

```python
## Where does Viterbi do better?
 
better_than_vit = data[ (data.fb_prune_sc - data.viterbi_sc) > 0 ]
better_than_vit_t = trues[ (trues.fb_prune_sc - trues.viterbi_sc) > 0 ]
print( "better_than_vit:", len(better_than_vit), len(better_than_vit_t), len(better_than_vit_t)/len(better_than_vit) )

worse_than_vit = data[ (data.fb_prune_sc - data.viterbi_sc) < 0 ]
worse_than_vit_t = trues[ (trues.fb_prune_sc - trues.viterbi_sc) < 0 ]
print( "worse_than_vit:", len(worse_than_vit), len(worse_than_vit_t), len(worse_than_vit_t)/len(worse_than_vit) )

worse_than_vit_t
```

```python
## Where does fb-pruner do much worse than Forward-Backward?
difference = 50
big_diff = data[ data.prune_below_fwd > difference ]
print("len:", len(big_diff))
big_diff.res_id
```

```python
## Large cases

data.total_cells.max()
cells_std_2above = data.total_cells.mean() + (data.total_cells.std() * 2)
cells_std_2above

# two std dev above normal
big_cases = data[ data.total_cells >= cells_std_2above ]
big_cases
```

```python
## Avg stats

names = [ "full_data", "trues", "falses", "worse_than_vit", "big_diff_from_fwd", "big_cases" ]
datasets = [ data, trues, falses, worse_than_vit, big_diff, big_cases ]

print("STATS:\n")

for i in range(len(datasets)):
    dataset = datasets[i]
    name = names[i]
    print( name.upper() )
    print( "total cells avg:\t\t", int(dataset.total_cells.mean()), int(dataset.total_cells.std()) )
    print( "calc'd cells avg:\t\t", int(dataset.calc_cells.mean()), int(dataset.calc_cells.std()) )
    print( "percent cells avg:\t\t", dataset.perc_cells.mean(), dataset.perc_cells.std() )
    print( "prune-fwd difference:\t\t", dataset.prune_below_fwd.mean(), dataset.prune_below_fwd.std() )
    print( "forward score:\t\t\t", dataset.forward_sc.mean(), dataset.forward_sc.std() )
    print( "fb-pruner score:\t\t", dataset.fb_prune_sc.mean(),  dataset.fb_prune_sc.std() )
    print( "viterbi score:\t\t\t", dataset.viterbi_sc.mean(),  dataset.viterbi_sc.std() )
    nt = len(dataset[dataset.tf == "T"])
    nf = len(dataset[dataset.tf == "F"])
    tf_ratio = nt/len(dataset)
    print( "t/f ratio:\t\t\t", tf_ratio, nt, nf )

#     print(data.mean())
#     print(data.std())
    print()
```

```python
## plot comparison
subdata = data[ data.forward_sc < 20 ]
subdata = subdata.sample( frac = 0.01 )

# get maximum and minimum data points
min_sc = min( [ min( subdata.fb_prune_sc ), min( subdata.viterbi_sc ), min( subdata.forward_sc ) ] )
max_sc = max( [ max( subdata.fb_prune_sc ), max( subdata.viterbi_sc ), max( subdata.forward_sc ) ] )
print( "min_sc: {:.2}, max_sc: {:.2}".format( min_sc, max_sc ) )

# full forward / perfect scores
m,b = np.polyfit( subdata.forward_sc, subdata.forward_sc, 1 )
print("full-forward: f(x) = {:.2}x + {:.2}".format(m,b))
fwd_fitline = [ m * min_sc + b, m * max_sc + b ] 
# plot line
plt.plot( fwd_fitline, fwd_fitline, color='lightgreen', label='Full Forward Score', linestyle='-', linewidth=1, alpha=1, zorder=4 )
plt.plot( fwd_fitline, fwd_fitline, color='black', linestyle='-', linewidth=3, alpha=1, zorder=3 )

# compare method to full forward scores
plt.scatter( subdata.fb_prune_sc, subdata.forward_sc, color='blue', label='Pruned Forward Score', marker='v', s=2, alpha=0.2, zorder=5 )
plt.scatter( subdata.viterbi_sc, subdata.forward_sc, color='red', label='Viterbi Score', marker='^', s=2, alpha=0.2, zorder=5 )

# method fit to subdata
m,b = np.polyfit( subdata.fb_prune_sc, subdata.forward_sc, 1 )
fb_prune_fitline = [ m * min_sc + b, m * max_sc + b ] 
print("fb-prune (subdata): f(x) = {:.2}x + {:.2}".format(m,b))
m,b = np.polyfit( subdata.viterbi_sc, subdata.forward_sc, 1 )
viterbi_fitline = [ m * min_sc + b, m * max_sc + b ]
print("viterbi (subdata): f(x) = {:.2}x + {:.2}".format(m,b))

# method fit to full data
m,b = np.polyfit( data.fb_prune_sc, data.forward_sc, 1 )
fb_prune_fitline = [ m * min_sc + b, m * max_sc + b ] 
print("fb-prune (data): f(x) = {:.2}x + {:.2}".format(m,b))
m,b = np.polyfit( data.viterbi_sc, data.forward_sc, 1 )
viterbi_fitline = [ m * min_sc + b, m * max_sc + b ]
print("viterbi (data): f(x) = {:.2}x + {:.2}".format(m,b))

# plot fitted lines
plt.plot( fwd_fitline, fb_prune_fitline, color='lightblue', label='Pruned Forward Fit', linestyle='-', linewidth=1, alpha=1, zorder=4 )
plt.plot( fwd_fitline, fb_prune_fitline, color='black', linestyle='-', linewidth=3, alpha=1, zorder=3 )
plt.plot( fwd_fitline, viterbi_fitline, color='lightcoral', label='Viterbi Fit', linestyle='-', linewidth=1, alpha=1, zorder=4 )
plt.plot( fwd_fitline, viterbi_fitline, color='black', linestyle='-', linewidth=3, alpha=1, zorder=3 )

plt.xlim( max_sc, min_sc )
plt.ylim( max_sc, min_sc )
# plt.yscale('log')
# plt.xscale('log')
plt.legend(loc='upper left')
plt.xlabel('Full Forward Score')
plt.ylabel('Method Score')
plt.title('Method Scores Relative to Forward Scores (Fwd Scores < 20)')
plt.grid(True)
plt.show()

plt.savefig("Scores-Relative-to-Forward-Scores_20.jpg")
```
