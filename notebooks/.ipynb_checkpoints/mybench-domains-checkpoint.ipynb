# Load Data

### Diff_Align Table

```python
datas = []

# set header format
diffaln_dformat = {
    "t_name": "str",
    "q_name": "str",
    "truth": "str",
    "mmseqs_sc": "float",
    "hmmer_sc": "float",
    "diff_sc": "float",
    "overlap?": "float",
    "t_cover": "float",
    "q_cover": "float",
    "cover": "float",
    "hmmer_t_beg": "int",
    "hmmer_t_end": "int",
    "mmseqs_t_beg": "int",
    "mmseqs_t_end": "int",
    "t_len": "int",
    "hmmer_q_beg": "int",
    "hmmer_q_end": "int",
    "mmseqs_q_beg": "int",
    "mmseqs_q_end": "int",
    "q_len": "int",
}
diffaln_header = diffaln_dformat.keys()

# load data
usecols = [0,1,2,3,4,5]
skiprows = []
sep = " "
names = diffaln_header
```

```python
diffaln_data = pd.read_csv( diffaln_fname, sep=sep, header=None, skiprows=skiprows, names=names, index_col=False, verbose=False )
datas.append(diffaln_data)
diffaln_data
```

# Plots

```python
safe_colors = ['#161925', '#EF5A39', '#602661', '#12c5a7', '#082a6a', '#cbcbcb', '#e926ef', '#fdcfb7' ]
safe_colors_pastel = ['', '', '', '', '', ]

bins = np.linspace(0,1,21)
print(bins)
```

```python
data = diffaln_data

cover    = data["cover"]
t_cover  = data["t_cover"]
q_cover  = data["q_cover"]

# plt.hist( t_cover, bins, color='red', alpha=.7, linewidth=1, edgecolor='black', label="Coverage of Target" )
# plt.hist( q_cover, bins, color='blue', alpha=.7, linewidth=1, edgecolor='black', label="Coverage of Query" )
plt.hist( q_cover * t_cover, bins, color='green', alpha=.7, linewidth=1, edgecolor='black', label="Coverage of Domain DP Matrix Region" )

plt.grid(color='gray', linestyle='-', linewidth=1, alpha=0.2)

plt.title('Coverage of True-Positive Domains by MMSEQS Relative to HMMER')
plt.ylabel('Number of Domains')
plt.xlabel('Fractional Coverage of HMMER-selected Domain')
plt.legend(loc='upper left')

plt.show()
```

```python
data = diffaln_data

cover    = data["cover"]
t_cover  = data["t_cover"]
q_cover  = data["q_cover"]

dp_size  = data["t_len"] * data["q_len"]

plt.scatter( dp_size, cover, color='green', alpha=.4, marker='^', s=1.5, label="Coverage of Domain DP Matrix Region" )

plt.grid(color='gray', linestyle='-', linewidth=1, alpha=0.2)

plt.title('Coverage of True-Positive Domains by MMSEQS Relative to HMMER')
plt.xlabel('Size of DP Matrix (Query x Target)')
plt.ylabel('Fraction Coverage of HMMER-selected Domain')
plt.xscale('log')

plt.show()
```

```python
data = diffaln_data

mmseqs_t_len = data["hmmer_t_end"] - data["hmmer_t_beg"]
hmmer_t_len  = data["mmseqs_t_end"] - data["mmseqs_t_beg"]

mmseqs_q_len = data["hmmer_q_end"] - data["hmmer_q_beg"]
hmmer_q_len  = data["mmseqs_q_end"] - data["mmseqs_q_beg"]

x_s = [0, 1500]
plt.plot( x_s, x_s, color='green', zorder=1 )

plt.scatter( hmmer_q_len, mmseqs_q_len, color='blue', alpha=.5, marker='^', s=1.0, label="Query Lengths", zorder=2 )
# plt.scatter( hmmer_t_len, mmseqs_t_len, color='red', alpha=.2, marker='^', s=1.0, label="Target Lengths", zorder=2 )

plt.grid(color='gray', linestyle='-', linewidth=1, alpha=0.2)
plt.xlabel('Length of HMMER Domain')
plt.ylabel('Length of MMSEQS Domain')
plt.ylim(0, 1000)
plt.xlim(0, 1000)

plt.show()
```
