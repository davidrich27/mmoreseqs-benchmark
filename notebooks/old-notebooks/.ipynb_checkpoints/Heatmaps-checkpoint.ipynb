```python
#imports
from __future__ import print_function
import sys
import os
import psutil
import math
import numpy as np
import pandas as pd
import cv2 as cv
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import statsmodels.api as sm
import statistics as stats

matplotlib.rcParams['figure.dpi'] = 300
pd.set_option('display.max_columns', None)  
```

```python
# index filepaths
target_index_fname = "/home/devreckas/Data/profmark-benchmark/temp/cloud/pmark.hmm.idx"
query_index_fname = "/home/devreckas/Data/profmark-benchmark/temp/cloud/pmark.fa.idx"
```

```python
# data filepaths
hmmer_generic_fname = "/home/devreckas/Data/profmark-benchmark/results/hmmer-test/hmmer-test-generic.7090.txt.tmp"
mmseqs_plus_fname = "/home/devreckas/Data/profmark-benchmark/results/cloud/final_03/final.result.16.mym8"
```

#### indexes

```python
# set header format
target_index_dformat = {
    "target_id": "int64",
    "target_offset": "int64",
    "target_name": "str"
}
query_index_dformat = {
    "query_id": "int64",
    "query_offset": "int64",
    "query_name": "str"
}
target_index_header = target_index_dformat.keys()
query_index_header = query_index_dformat.keys()
```

```python
# load target data
usecols = [0,2]
skiprows = None
sep = "\t"
names = target_index_header
target_index = pd.read_csv( target_index_fname, sep=sep, header=None, skiprows=skiprows, usecols=usecols, names=names, index_col=False, verbose=False )
target_index
```

```python
# load query data
usecols = [0,2]
skiprows = None
sep = "\t"
names = query_index_header
query_index = pd.read_csv( query_index_fname, sep=sep, header=None, skiprows=skiprows, usecols = usecols, names=names, index_col=False, verbose=False )
query_index
```

#### HMMER generic v. optimized data

```python
# set header names and datatype
hmmer_generic_dformat = {
    "header_id": "str", 
    "target_length": "int64",
    "target_name": "str",
    "query_length": "int64", 
    "query_name": "str", 
    "forward_gen_time": "float64", 
    "forward_gen_score": "float64", 
    "forward_opt_time": "float64",
    "forward_opt_score": "float64",
    "sequence_score": "float64"
}
hmmer_generic_header = hmmer_generic_dformat.keys()
```

```python
# load data
sep = " "
usecols = [1,2,3,4,5,6,7,8]
skiprows = [0,1]
names = hmmer_generic_header
hmmer_generic_data = pd.read_csv( hmmer_generic_fname, sep=" ", header=None, skiprows=skiprows, usecols = usecols, names=names, index_col=False, verbose=False )
hmmer_generic_data
```

```python
# cast data to proper datatypes
cols = list(hmmer_generic_data.columns.values)
for i in range(len(cols)):
    col = cols[i]
    dtype = hmmer_generic_dformat[col]
    hmmer_generic_data.iloc[:,i] = hmmer_generic_data.iloc[:,i].astype(dtype)
hmmer_generic_data.dtypes
```

```python
# add target and query id's
key = 'target_name'
hmmer_generic_data = hmmer_generic_data.merge( target_index, how='inner', left_on=key, right_on=key)
key = 'query_name'
hmmer_generic_data = hmmer_generic_data.merge( query_index, how='inner', left_on=key, right_on=key)
hmmer_generic_data.dtypes
```

#### MMSEQS Plus data

```python
# set header names and datatype
mmseqs_plus_dformat = {
    "result_id": "int64", 
    "target_id": "int64", 
    "query_id": "int64", 
    "target_length": "int64", 
    "query_length": "int64", 
    "total_cells": "int64", 
    "cloud_cells": "int64", 
    "alpha": "float64", 
    "beta": "int64", 
    "viterbi_score": "float64", 
    "forward_score": "float64", 
    "backward_score": "float64", 
    "cloud_score": "float64", 
    "viterbi_time": "float64", 
    "forward_time": "float64", 
    "backward_time": "float64", 
    "cloud_forward_time": "float64", 
    "cloud_backward_time": "float64", 
    "merge_time": "float64", 
    "reorient_time": "float64", 
    "bound_forward_time": "float64", 
    "cloud_total_time": "float64"
}
mmseqs_plus_header = mmseqs_plus_dformat.keys()
```

```python
# load mmseqs data 
sep = "\t"
usecols = range(len(mmseqs_plus_dformat))
skiprows = [0,1]
names = mmseqs_plus_header
mmseqs_plus_data = pd.read_csv( mmseqs_plus_fname, sep=sep, header=None, skiprows=skiprows, usecols=usecols, names=names, index_col=False, verbose=False )
# clean improperly formatted rows
mmseqs_plus_data.result_id = pd.to_numeric(mmseqs_plus_data.result_id, errors='coerce')
mmseqs_plus_data = mmseqs_plus_data[mmseqs_plus_data.result_id.notnull()]
mmseqs_plus_data
```

```python
# cast data to proper datatypes
keys = list(mmseqs_plus_data.columns.values)
for i in range(len(keys)):
    key = keys[i]
    dtype = mmseqs_plus_dformat[key]
    mmseqs_plus_data.iloc[:,i] = mmseqs_plus_data.iloc[:,i].astype(dtype)
mmseqs_plus_data.dtypes
```

#### Join data

```python
# inner join on target and query id's
key=['target_id', 'query_id', 'target_length', 'query_length']
joint_data = hmmer_generic_data.merge( mmseqs_plus_data, left_on=key, right_on=key)
# add total size
total_cells = ( joint_data.target_length+1 ) * ( joint_data.query_length+1 )
joint_data["total_cells"] = total_cells
joint_data
```
